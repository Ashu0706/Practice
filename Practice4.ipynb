```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "department": ["IT", "IT", "HR", "HR","MR","MR"],
    "age": [25, np.nan, 30, 22, 34,34],
    "salary": [50000, 60000, np.nan, 45000,2000,2000]
})
df
```

```python
# Drop rows with missing values
df_drop = df.dropna()
df_drop
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df
```

```python
df['salary'].fillna(df['salary'].mean(),inplace=True)
df
```

```python
df[df.duplicated() == False]
```

```python
df.groupby('department')['salary'].sum()
```

```python
df['salary'] = df['salary']*2
df
```

```python
df['runnuing_total'] = df['salary'].cumsum()
df
```
