```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('practice6').getOrCreate()
```

```python
schema = ('customer_id','product_key')
cust_data =[(1,5),
            (2,6),
            (3,5),
            (3,6),
            (1,6),
            (2,6)]
cust_df = spark.createDataFrame(cust_data,schema)
cust_df.show()

prod_df = spark.createDataFrame([(5,), (6,)], ['product_key'])
prod_df.show()
```

```python
from pyspark.sql.functions import count,col,count_distinct,countDistinct
c = prod_df.agg(count('product_key')).collect()[0][0]
res = cust_df.groupBy(col('customer_id')).agg(countDistinct(col('product_key')).alias('cnt')).filter(col('cnt') == c)
res.show()
```
